```python
import warnings
from statsmodels.tools.sm_exceptions import ValueWarning
warnings.filterwarnings("ignore", category=ValueWarning)
warnings.filterwarnings("ignore")
print("Done!")

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import zipfile # Import zipfile for data extraction
import warnings
from statsmodels.tools.sm_exceptions import ValueWarning

warnings.filterwarnings("ignore", category=ValueWarning)
warnings.filterwarnings("ignore")

from sklearn.base import BaseEstimator, TransformerMixin, RegressorMixin, clone
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, FunctionTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import joblib # Keep joblib for local model saving
from tqdm import tqdm
from neuralforecast import NeuralForecast
from neuralforecast.models import TFT
pd.set_option("display.max_rows", 100)



```

```python
def calculate_wmae(y_true, y_pred, is_holiday_flag, holiday_weight=5.0):

    abs_errors = np.abs(y_true - y_pred)
    weights = np.where(is_holiday_flag.astype(bool), holiday_weight, 1.0)
    wmae = np.sum(weights * abs_errors) / np.sum(weights)
    return wmae
```

```python
class DateFeatureCreator(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        if "Date" not in X.columns:
            # This should not happen if MultiIndexKeeper precedes it and drop=False
            raise ValueError("DateFeatureCreator requires 'Date' column in input X.")

        X["week"] = (X["Date"].dt.to_period("W").rank(method="dense").astype(int) - 1)
        X["sin_13"] = np.sin(2 * np.pi * X["week"] / 13)
        X["cos_13"] = np.cos(2 * np.pi * X["week"] / 13)
        X["sin_23"] = np.sin(2 * np.pi * X["week"] / 23)
        X["cos_23"] = np.cos(2 * np.pi * X["week"] / 23)

        # Drop 'Date' column after feature creation
        # It's okay to drop it here because MultiIndexKeeper ensures it's in the index,
        # and TFTRegressor extracts it from the index or from the `df` it receives.
        X = X.drop(columns=["Date"], errors='ignore')
        return X
```

```python
class ColumnDropper(BaseEstimator, TransformerMixin):

    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X.drop(columns=self.columns, errors="ignore")
```

```python
class ColumnTransformerWithNames(ColumnTransformer):
    def get_feature_names_out(self, input_features=None):
        feature_names = []
        for name, transformer, columns_in_transformer in self.transformers_:
            if transformer == 'drop':
                continue
            elif transformer == 'passthrough':
                feature_names.extend(columns_in_transformer)
                continue

            # Ensure columns_in_transformer is always a list for consistent iteration
            if isinstance(columns_in_transformer, str):
                columns_in_transformer = [columns_in_transformer]

            if hasattr(transformer, 'get_feature_names_out'):
                try:
                    # For transformers that accept input_features (e.g., OneHotEncoder in newer scikit-learn)
                    sub_feature_names = transformer.get_feature_names_out(columns_in_transformer)
                except (TypeError, ValueError):
                    # Fallback for older scikit-learn or transformers that don't take input_features
                    sub_feature_names = transformer.get_feature_names_out()
                    # Manually handle OneHotEncoder if get_feature_names_out() doesn't give full names
                    if isinstance(transformer, OneHotEncoder):
                        # Construct names based on categories_ attribute
                        ohe_generated_names = []
                        for i, col_name in enumerate(columns_in_transformer):
                            categories = transformer.categories_[i]
                            for cat in categories:
                                ohe_generated_names.append(f"{col_name}_{cat}")
                        sub_feature_names = np.array(ohe_generated_names)
                    else:
                        # Fallback if get_feature_names_out does not provide specific names
                        sub_feature_names = np.array([f"{name}__{col}" for col in columns_in_transformer])
            else:
                # If transformer doesn't have get_feature_names_out (e.g., SimpleImputer)
                sub_feature_names = np.array([f"{col}" for col in columns_in_transformer]) # No prefix needed often

            feature_names.extend(sub_feature_names)
        return np.array(feature_names)

    def transform(self, X):
        X_transformed = super().transform(X)
        # Use X.columns for input_features to get_feature_names_out
        cols = self.get_feature_names_out(X.columns)

        cleaned_cols = []
        for col_name in cols:
            if '__' in col_name:
                cleaned_cols.append(col_name.split("__", 1)[-1])
            else:
                cleaned_cols.append(col_name)

        res = pd.DataFrame(X_transformed, columns=cleaned_cols, index=X.index)
        return res

    def fit_transform(self, X, y=None):
        X_transformed = super().fit_transform(X, y)
        cols = self.get_feature_names_out(X.columns) # Pass X.columns here

        cleaned_cols = []
        for col_name in cols:
            if '__' in col_name:
                cleaned_cols.append(col_name.split("__", 1)[-1])
            else:
                cleaned_cols.append(col_name)

        res = pd.DataFrame(X_transformed, columns=cleaned_cols, index=X.index)
        return res
```

```python
class MultiIndexKeeper(BaseEstimator, TransformerMixin):
    def __init__(self, index_cols=["Date", "Store", "Dept"]):
        self.index_cols = index_cols

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        if 'Date' in X.columns and not pd.api.types.is_datetime64_any_dtype(X['Date']):
            X['Date'] = pd.to_datetime(X['Date'])

        missing_cols = [col for col in self.index_cols if col not in X.columns]
        if missing_cols:
            raise ValueError(f"MultiIndexKeeper: Missing columns in input X: {missing_cols}")

        X.set_index(self.index_cols, drop=False, inplace=True)
        return X
```

```python
class TFTRegressor(BaseEstimator, RegressorMixin):
    def __init__(self, input_chunk_length=52, output_chunk_length=39, epochs=10, batch_size=32, random_seed=42):
        self.input_chunk_length = input_chunk_length
        self.output_chunk_length = output_chunk_length
        self.epochs = epochs
        self.batch_size = batch_size
        self.random_seed = random_seed
        self.nf_ = None
        self.model_ = None

    def fit(self, X, y):
        required_cols = ["Store", "Dept", "Date"]
        for col in required_cols:
            # Check in columns or index levels
            if col not in X.columns and col not in X.index.names:
                raise ValueError(f"TFTRegressor.fit: Missing required column/index '{col}' in X.")

        df = X.copy()
        df["y"] = y.values

        if 'Date' in df.index.names:
            df = df.reset_index(level='Date')
        if not pd.api.types.is_datetime64_any_dtype(df['Date']):
            df['Date'] = pd.to_datetime(df['Date'])

        df.rename(columns={"Date": "ds"}, inplace=True)
        df["unique_id"] = df["Store"].astype(str) + "_" + df["Dept"].astype(str)



        self.model_ = TFT(
            h=self.output_chunk_length,
            input_size=self.input_chunk_length,
            batch_size=self.batch_size,
            random_seed=self.random_seed,
        )

        self.nf_ = NeuralForecast(models=[self.model_], freq="W-FRI")
        self.nf_.fit(df=df)

        return self
```

```python
def predict(self, X):
        df_predict = X.copy()

        if 'Date' not in df_predict.columns and 'Date' in X.index.names:
            df_predict = df_predict.reset_index(level='Date')

        if not pd.api.types.is_datetime64_any_dtype(df_predict['Date']):
            df_predict['Date'] = pd.to_datetime(df_predict['Date'])

        df_predict.rename(columns={"Date": "ds"}, inplace=True)
        df_predict["unique_id"] = df_predict["Store"].astype(str) + "_" + df_predict["Dept"].astype(str)

        forecast_df = self.nf_.predict()

        model_name = self.model_.__class__.__name__
        if model_name in forecast_df.columns:
            forecast_df = forecast_df.rename(columns={model_name: "y_hat"})
        else:
            raise ValueError(f"Expected prediction column '{model_name}' not found in forecast_df. Columns: {forecast_df.columns}")

        temp_df_for_merge = df_predict[['unique_id', 'ds']].copy()
        temp_df_for_merge['original_index'] = X.index

        merged = pd.merge(temp_df_for_merge, forecast_df[['unique_id', 'ds', 'y_hat']], on=["unique_id", "ds"], how="left")

        merged.set_index('original_index', inplace=True)

        preds = merged["y_hat"].fillna(0).values

        return preds.flatten()
```

```python
def main():
    print("🚀 Starting Walmart Sales Forecasting with Temporal Fusion Transformer")
    print("=" * 80)

    # --- 1. Data Loading ---
    print("\n📦 Step 1: Data Loading")
    # Using .zip directly to read compressed CSVs as pandas supports this
    try:
        df_train = pd.read_csv("train.csv", parse_dates=["Date"])
        df_features = pd.read_csv("features.csv", parse_dates=["Date"])
        df_stores = pd.read_csv("stores.csv")
        print("   ✅ Datasets loaded successfully.")
    except FileNotFoundError as e:
        print(f"❌ Error loading data: {e}. Please ensure the files are in the specified Kaggle input directory.")
        print("Expected paths: /kaggle/input/walmart-recruiting-store-sales-forecasting/train.csv.zip, etc.")
        return # Exit if files not found

    # --- 2. Data Merging and Initial Cleaning ---
    print("\n🔄 Step 2: Data Merging and Initial Cleaning")
    # Merge datasets
    df = df_train.merge(df_features, on=["Store", "Date", "IsHoliday"], how="left")
    df = df.merge(df_stores, on="Store", how="left")

    # Handle NaN values for markdown features (assume 0 if no markdown)
    markdown_cols = [f'MarkDown{i}' for i in range(1, 6)]
    for col in markdown_cols:
        df[col] = df[col].fillna(0)

    # Filter out negative sales in training data
    initial_rows = len(df)
    df = df[df['Weekly_Sales'] >= 0]
    if len(df) < initial_rows:
        print(f"   🗑️ Removed {initial_rows - len(df)} rows with negative Weekly_Sales.")

    # Sort data by Date, Store, Dept for time-based splitting and TFT
    df = df.sort_values(by=['Date', 'Store', 'Dept']).reset_index(drop=True)
    print(f"   ✅ Data merged and cleaned. Final shape: {df.shape}")

    # --- 3. Data Splitting (80/20 Time-based) ---
    print("\n✂️ Step 3: Data Splitting (80/20 Time-based)")
    y = df["Weekly_Sales"]
    X = df.drop(columns=["Weekly_Sales"])

    # Calculate split index based on 80% of the data
    split_idx = int(len(X) * 0.8)
    X_train, X_val = X.iloc[:split_idx], X.iloc[split_idx:]
    y_train, y_val = y.iloc[:split_idx], y.iloc[split_idx:]

    print(f"   Train set shape: {X_train.shape}, Validation set shape: {X_val.shape}")
    print(f"   Train target shape: {y_train.shape}, Validation target shape: {y_val.shape}")
    print("   ✅ Data split complete.")

    # --- 4. Pipeline Definition ---
    print("\n⚙️ Step 4: Pipeline Definition")

    date_features = ["week", "sin_13", "cos_13", "sin_23", "cos_23"]
    engineered_features = date_features

    # IMPORTANT CHANGE HERE:
    # 'Store', 'Dept' should NOT be passed to the OneHotEncoder if TFTRegressor uses them directly
    # for unique_id creation. 'Type' and 'IsHoliday' can still be one-hot encoded as covariates.

    # These columns will be passed through unchanged to TFTRegressor
    passthrough_cols = ["Store", "Dept"]

    # These will be one-hot encoded
    categorical_ohe_cols = ["Type", "IsHoliday"]

    numerical_cols = ["Temperature", "Fuel_Price", "CPI", "Unemployment",
                      "MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]

    categorical_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])
    numerical_transformer = SimpleImputer(strategy="mean")

    preprocessor = ColumnTransformerWithNames(
        transformers=[
            ("num", numerical_transformer, numerical_cols + engineered_features),
            ("cat", categorical_transformer, categorical_ohe_cols), # Only Type and IsHoliday now
            ("passthrough_ids", "passthrough", passthrough_cols) # Pass Store and Dept through
        ],
        sparse_threshold=0.0
    )

    pipeline = Pipeline([
        ("index_keeper", MultiIndexKeeper()), # Keeps Date, Store, Dept in index and as columns
        ("date_features", DateFeatureCreator()), # Creates cyclical features and drops 'Date' from columns
        ("preprocessor", preprocessor), # Processes other numerical/categorical features
        ("tft_regressor", TFTRegressor(input_chunk_length=52, output_chunk_length=39, epochs=25, batch_size=32))
    ])
    print("   ✅ Pipeline defined.")

    # --- 5. Model Training ---
    print("\n🚀 Step 5: Training the TFT Model")
    print("   This may take some time depending on data size and hardware.")
    try:
        pipeline.fit(X_train, y_train)
        best_model = pipeline # In this setup, the last fitted pipeline is the "best_model"
        print("   ✅ TFT model training finished.")
    except Exception as e:
        print(f"❌ Error during model training: {e}")
        return

    # --- 6. Model Evaluation ---
    print("\n📈 Step 6: Evaluating Model Performance on Validation Set")
    try:
        preds_val = best_model.predict(X_val) # Predict on the validation set

        # Ensure predictions are non-negative
        preds_val[preds_val < 0] = 0

        # Calculate MAE
        mae = mean_absolute_error(y_val, preds_val)
        print(f"   Validation MAE: {mae:.4f}")


        is_holiday_flag_val = X_val['IsHoliday'].astype(int) # This assumes X_val still contains original 'IsHoliday'

        wmae = calculate_wmae(y_val.to_numpy(), preds_val, is_holiday_flag_val.to_numpy())
        print(f"   Validation WMAE: {wmae:.4f}")
        print("   ✅ Model evaluation complete.")

    except Exception as e:
        print(f"❌ Error during model evaluation: {e}")
        return

    print("\n🎉 Walmart Sales Forecasting with TFT (Training & Validation Only) - COMPLETE!")
    print("=" * 80)
```
